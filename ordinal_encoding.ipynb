```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('customer.csv')
df
```

```python
df.info()
```

```python
df = df.iloc[:, 2:] 
#age is already numeric
#and gender is nominal data
#removing age and gender to apply ordinal encoder on review and education
#label encoder on purchased - y
df.head()
```

```python
X = df.iloc[:, 0:2]
X
```

```python
y = df.iloc[:, -1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder

#get ordinal encoder and also define categories
oe = OrdinalEncoder(categories=[["Poor", "Average", "Good"], ["School", "UG", "PG"]])

#fit ordinal encoder
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(y_train) #use label encoder on y - purchased
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
y_train
```
